```python
# ============================================================
# Colab: Hyperbolic clustering + Poincaré disk visualization
# (WordNet-style disk with induced edges + cluster listings)
#
# Changes vs previous:
# - Clustering moved to tangent space: T = logmap0(X_ball)
#   -> use BisectingKMeans (more balanced) OR Ward
# - Optional k selection via silhouette in tangent space
# - Plot labels are NAME ONLY (no language code)
# ============================================================

!pip -q install scipy umap-learn adjustText

import os, re, glob, math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection

from google.colab import drive
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering, BisectingKMeans
from sklearn.metrics import silhouette_score
from scipy.sparse.csgraph import minimum_spanning_tree

# -----------------------
# 1) User config
# -----------------------
BASE_DIR = "/content/drive/MyDrive/Phd/datasets/wav2vec2_embeddings/paper1"
NPZ_GLOB = os.path.join(BASE_DIR, "*.npz")

LAYER_BINS = [(0,4), (5,9), (10,14), (15,19), (20,24)]
ALL_LAYERS = list(range(0, 25))

# Hyperbolic parameters
C = 1.0
R_TARGET_BALL = 0.60
R_TARGET_VIS = 0.60

# --- Clustering config (balanced) ---
CLUSTER_METHOD = "bisectkmeans"   # "bisectkmeans" | "ward"
N_CLUSTERS = 10                   # target k (used if AUTO_K=False)
AUTO_K = False                    # if True, choose k by silhouette in tangent PCA space
K_RANGE = list(range(4, 21))      # used if AUTO_K=True
TANGENT_PCA_DIM_FOR_CLUSTERING = 32  # reduce dims for stability/speed (tangent space)

# Graph edges for visualization
EDGE_MODE = "mst"  # "mst" | "knn"
KNN_K = 3

# Plotting
FIGSIZE = (10, 10)
POINT_SIZE = 18
EDGE_ALPHA = 0.18
EDGE_LW = 0.20

# Label options (names only)
LABEL_MODE = "subset"     # "none" | "subset" | "all"
LABEL_TOP_K_PER_CLUSTER = 6
LABEL_FONT_SIZE = 8

PRINT_CLUSTER_LISTS = True

EPS = 1e-9
SAFE_EPS_BALL = 1e-5

# -----------------------
# 2) Mount Drive
# -----------------------
drive.mount("/content/drive")

# ------------------------------------------------------------
# 3) Filename parsing
# ------------------------------------------------------------
LANG_RE = re.compile(r"embeddings_language-(.+?)_languagecode-\[([^\]]+)\]")

def parse_lang_from_filename(path: str):
    fn = os.path.basename(path)
    m = LANG_RE.search(fn)
    if m:
        return m.group(1), m.group(2)
    return os.path.splitext(fn)[0], "unknown"

# ------------------------------------------------------------
# 4) Robust layer extractor
# ------------------------------------------------------------
def get_layer_matrix(npz: np.lib.npyio.NpzFile, layer: int) -> np.ndarray:
    keys = list(npz.keys())

    for k in ["embeddings", "X", "hidden_states", "layer_embeddings"]:
        if k in keys:
            arr = np.asarray(npz[k])
            if arr.ndim == 3 and arr.shape[0] > layer and arr.shape[1] >= 1:
                return arr[layer]           # (N,D)
            if arr.ndim == 3 and arr.shape[1] > layer and arr.shape[0] >= 1:
                return arr[:, layer, :]     # (N,D)
            if arr.ndim == 2:
                return arr                  # (N,D)
            raise ValueError(f"Unrecognized shape for key '{k}': {arr.shape}")

    candidates = [
        f"embeddings_layer_{layer}",
        f"layer_{layer}",
        f"hidden_state_{layer}",
        f"h_{layer}",
        f"layer{layer}",
    ]
    for ck in candidates:
        if ck in keys:
            arr = np.asarray(npz[ck])
            if arr.ndim != 2:
                raise ValueError(f"Key '{ck}' exists but is not 2D: shape={arr.shape}")
            return arr

    layer_pat = re.compile(rf"(^|[^0-9]){layer}([^0-9]|$)")
    for k in keys:
        if layer_pat.search(k):
            arr = np.asarray(npz[k])
            if arr.ndim == 2:
                return arr

    raise KeyError(f"Could not find embeddings for layer={layer}. keys={keys[:40]}")

# ------------------------------------------------------------
# 5) Load language: centroid per layer
# ------------------------------------------------------------
def load_language(npz_path: str, layers=ALL_LAYERS):
    lang_name, lang_code = parse_lang_from_filename(npz_path)
    z = np.load(npz_path, allow_pickle=True)

    centroids = {}
    for l in layers:
        X = get_layer_matrix(z, l)  # (N,D)
        X = np.asarray(X, dtype=np.float32)
        centroids[l] = X.mean(axis=0)

    z.close()
    return {"path": npz_path, "lang_name": lang_name, "lang_code": lang_code, "centroids": centroids}

paths = sorted(glob.glob(NPZ_GLOB))
if not paths:
    raise FileNotFoundError(f"No .npz files found under: {BASE_DIR}")

languages = []
for p in paths:
    try:
        languages.append(load_language(p))
    except Exception as e:
        print(f"[WARN] Skipping {os.path.basename(p)} due to: {e}")

print(f"Loaded {len(languages)} languages.")

# ------------------------------------------------------------
# 6) Identify Faroese (visualization centering only)
# ------------------------------------------------------------
def is_faroese(lang_name, lang_code):
    ln = (lang_name or "").lower()
    lc = (lang_code or "").lower().strip()
    return ("faro" in ln) or (lc in ["fo", "fo_fo", "fo-fo"]) or ("fao" in ln)

fo_candidates = [x for x in languages if is_faroese(x["lang_name"], x["lang_code"])]
if not fo_candidates:
    raise ValueError("Could not identify Faroese file. Adjust is_faroese().")
FO = fo_candidates[0]
print("Using Faroese:", FO["lang_name"], FO["lang_code"], os.path.basename(FO["path"]))

# ------------------------------------------------------------
# 7) Hyperbolic primitives
# ------------------------------------------------------------
def expmap0(v: np.ndarray, c: float, eps: float = 1e-9) -> np.ndarray:
    v = np.asarray(v, dtype=np.float64)
    norm = np.linalg.norm(v, axis=-1, keepdims=True)
    norm = np.clip(norm, eps, None)
    sqrt_c = math.sqrt(c)
    z = np.clip(sqrt_c * norm, 0.0, 20.0)
    factor = np.tanh(z) / np.clip(z, eps, None)
    return factor * v

def logmap0(x: np.ndarray, c: float, eps: float = 1e-9) -> np.ndarray:
    x = np.asarray(x, dtype=np.float64)
    norm = np.linalg.norm(x, axis=-1, keepdims=True)
    norm = np.clip(norm, eps, None)
    sqrt_c = math.sqrt(c)
    z = np.clip(sqrt_c * norm, 0.0, 1.0 - 1e-6)
    factor = np.arctanh(z) / (sqrt_c * norm)
    return factor * x

def poincare_dist(x: np.ndarray, y: np.ndarray, c: float, eps: float = 1e-9) -> float:
    x = np.asarray(x, dtype=np.float64)
    y = np.asarray(y, dtype=np.float64)
    x2 = float(np.dot(x, x))
    y2 = float(np.dot(y, y))
    max_r2 = (1.0 / c) - eps
    x2 = min(x2, max_r2)
    y2 = min(y2, max_r2)
    diff2 = float(np.dot(x - y, x - y))
    denom = (1.0 - c * x2) * (1.0 - c * y2)
    denom = max(denom, eps)
    z = 1.0 + (2.0 * c * diff2) / denom
    z = max(z, 1.0 + eps)
    return (1.0 / math.sqrt(c)) * math.acosh(z)

def mobius_add(x: np.ndarray, y: np.ndarray, c: float, eps: float = 1e-9) -> np.ndarray:
    x = np.asarray(x, dtype=np.float64)
    y = np.asarray(y, dtype=np.float64)
    x2 = np.sum(x*x, axis=-1, keepdims=True)
    y2 = np.sum(y*y, axis=-1, keepdims=True)
    xy = np.sum(x*y, axis=-1, keepdims=True)
    num = (1 + 2*c*xy + c*y2) * x + (1 - c*x2) * y
    den = 1 + 2*c*xy + (c**2) * x2 * y2
    den = np.clip(den, eps, None)
    return num / den

def mobius_translate_to_origin(X: np.ndarray, x0: np.ndarray, c: float) -> np.ndarray:
    return mobius_add(-x0, X, c=c, eps=EPS)

# ------------------------------------------------------------
# 8) Auto-scaling
# ------------------------------------------------------------
def choose_alpha_for_target_radius(Vz: np.ndarray, c: float, r_target: float = 0.6, eps: float = 1e-9) -> float:
    norms = np.linalg.norm(Vz.astype(np.float64), axis=1)
    med = float(np.median(norms))
    med = max(med, eps)
    sqrt_c = math.sqrt(c)
    r_cap = min(r_target, (1.0 - 1e-5) / sqrt_c)
    t_target = (1.0 / sqrt_c) * math.atanh(r_cap * sqrt_c)
    return t_target / med

def sat_fraction(X_ball: np.ndarray, c: float, thresh: float = 0.95) -> float:
    R = 1.0 / math.sqrt(c)
    r = np.linalg.norm(X_ball, axis=1)
    return float(np.mean(r >= thresh * R))

def hyperbolic_points_from_vectors(Vz: np.ndarray, alpha: float, c: float) -> np.ndarray:
    X = expmap0(alpha * Vz, c=c, eps=EPS)
    r2 = np.sum(X*X, axis=1)
    rmax = (1.0 - SAFE_EPS_BALL)**2 / c
    too_big = r2 >= rmax
    if np.any(too_big):
        scale = np.sqrt(rmax / np.clip(r2[too_big], 1e-12, None))
        X[too_big] *= scale[:, None]
    return X

def poincare_disk_coords_scaled(X_ball: np.ndarray, c: float, r_target: float = 0.6) -> np.ndarray:
    T = logmap0(X_ball, c=c, eps=EPS)
    T2 = PCA(n_components=2, random_state=0).fit_transform(T)
    norms2 = np.linalg.norm(T2, axis=1)
    med2 = float(np.median(norms2))
    med2 = max(med2, 1e-9)
    sqrt_c = math.sqrt(c)
    r_cap = min(r_target, (1.0 - 1e-5) / sqrt_c)
    t_target = (1.0 / sqrt_c) * math.atanh(r_cap * sqrt_c)
    vis_scale = t_target / med2
    return expmap0(T2 * vis_scale, c=c, eps=EPS)

# ------------------------------------------------------------
# 9) Poincaré distance matrix (for edges / sanity only)
# ------------------------------------------------------------
def pairwise_poincare_distance_matrix(X_ball: np.ndarray, c: float) -> np.ndarray:
    n = X_ball.shape[0]
    D = np.zeros((n, n), dtype=np.float64)
    for i in range(n):
        for j in range(i+1, n):
            d = poincare_dist(X_ball[i], X_ball[j], c=c, eps=EPS)
            D[i, j] = d
            D[j, i] = d
    return D

# ------------------------------------------------------------
# 10) Build bin centroids and standardize per bin across languages
# ------------------------------------------------------------
def bin_name(a, b): return f"L{a:02d}-{b:02d}"

BIN_CENTROIDS = {bin_name(a,b): [] for (a,b) in LAYER_BINS}
META = []  # (name, code, path)

for entry in languages:
    META.append((entry["lang_name"], entry["lang_code"], entry["path"]))
    for (a,b) in LAYER_BINS:
        mus = np.stack([entry["centroids"][l] for l in range(a, b+1)], axis=0)
        BIN_CENTROIDS[bin_name(a,b)].append(mus.mean(axis=0))

for k in BIN_CENTROIDS:
    BIN_CENTROIDS[k] = np.stack(BIN_CENTROIDS[k], axis=0).astype(np.float32)

BIN_CENTROIDS_Z = {}
for k, M in BIN_CENTROIDS.items():
    BIN_CENTROIDS_Z[k] = StandardScaler().fit_transform(M).astype(np.float32)

# Faroese index (for visualization centering only)
fo_idx_list = [i for i,(n,code,_) in enumerate(META) if code == FO["lang_code"]]
if not fo_idx_list:
    raise ValueError("Faroese not found after loading; check parsing.")
fo_idx = fo_idx_list[0]

# labels: NAME ONLY (requested)
labels_name_only = [n for (n,_,_) in META]

# ------------------------------------------------------------
# 11) Induced edges
# ------------------------------------------------------------
def edges_from_knn(D: np.ndarray, k: int = 3):
    N = D.shape[0]
    edges = set()
    for i in range(N):
        nn = np.argpartition(D[i], kth=min(k+1, N-1))[:k+1]
        nn = [j for j in nn if j != i][:k]
        for j in nn:
            a, b = (i, j) if i < j else (j, i)
            edges.add((a, b))
    return sorted(edges)

def edges_from_mst(D: np.ndarray):
    T = minimum_spanning_tree(D).tocoo()
    edges = set()
    for i, j in zip(T.row, T.col):
        a, b = (int(i), int(j)) if i < j else (int(j), int(i))
        edges.add((a, b))
    return sorted(edges)

# ------------------------------------------------------------
# 12) Cluster printing + label selection
# ------------------------------------------------------------
def print_cluster_lists(bin_str: str, clusters: np.ndarray, meta: list, max_per_cluster: int | None = None):
    print(f"\n--- Cluster membership (bin {bin_str}) ---")
    for cid in sorted(np.unique(clusters).tolist()):
        idx = np.where(clusters == cid)[0].tolist()
        items = [f"{meta[i][0]} [{meta[i][1]}]" for i in idx]  # list retains codes for text output
        if max_per_cluster is not None and len(items) > max_per_cluster:
            head = items[:max_per_cluster]
            print(f"Cluster {cid:>2}: n={len(items):>3} | " + ", ".join(head) + f", ... (+{len(items)-max_per_cluster})")
        else:
            print(f"Cluster {cid:>2}: n={len(items):>3} | " + ", ".join(items))

def select_label_indices(clusters: np.ndarray, fo_idx: int, top_k_per_cluster: int = 6):
    keep = {fo_idx}
    for cid in sorted(np.unique(clusters).tolist()):
        idx = np.where(clusters == cid)[0]
        for i in idx[:top_k_per_cluster]:
            keep.add(int(i))
    return sorted(keep)

# ------------------------------------------------------------
# 13) Plot
# ------------------------------------------------------------
def plot_poincare_disk_graph(coords2d_unit, clusters, fo_idx, title, edges=None,
                            labels=None, label_indices=None, show_labels=False,
                            epoch_text=None):
    coords2d_unit = np.asarray(coords2d_unit)
    fig, ax = plt.subplots(figsize=FIGSIZE)
    ax.set_aspect("equal")

    ax.add_patch(
        plt.Circle((0,0), 1.0, fill=True, ec="black", fc="#f8f9f9",
                   linestyle="-.", linewidth=1.0, zorder=0)
    )

    if edges is not None and len(edges) > 0:
        segs = [(coords2d_unit[i], coords2d_unit[j]) for (i, j) in edges]
        lc = LineCollection(segs, colors="blue", linewidths=EDGE_LW, alpha=EDGE_ALPHA, zorder=1)
        ax.add_collection(lc)

    ax.scatter(coords2d_unit[:,0], coords2d_unit[:,1],
               s=POINT_SIZE, c=clusters, alpha=0.90, zorder=2)

    ax.scatter([coords2d_unit[fo_idx,0]], [coords2d_unit[fo_idx,1]],
               marker="*", s=320, c="red", zorder=3, label="Faroese (centered)")

    if show_labels and labels is not None:
        if label_indices is None:
            label_indices = list(range(len(labels)))
        for i in label_indices:
            ax.text(coords2d_unit[i,0], coords2d_unit[i,1], labels[i],
                    fontsize=LABEL_FONT_SIZE, zorder=4)

    if epoch_text is not None:
        ax.text(-0.98, 0.90, epoch_text, fontsize=18, weight="bold", alpha=0.6)

    ax.set_title(title)
    ax.set_xlim([-1.05, 1.05])
    ax.set_ylim([-1.05, 1.05])
    ax.axis("off")
    ax.legend(loc="upper right")
    plt.show()

# ------------------------------------------------------------
# 14) Clustering in tangent space (balanced)
# ------------------------------------------------------------
def cluster_in_tangent_space(T: np.ndarray, method: str, k: int) -> np.ndarray:
    """
    T: (N,D) tangent vectors (Euclidean)
    """
    if method == "bisectkmeans":
        model = BisectingKMeans(n_clusters=k, random_state=0)
        return model.fit_predict(T)
    if method == "ward":
        # Ward tends to create compact, often more even clusters in Euclidean space
        model = AgglomerativeClustering(n_clusters=k, linkage="ward")
        return model.fit_predict(T)
    raise ValueError(f"Unknown method: {method}")

def choose_k_by_silhouette(T: np.ndarray, k_range: list[int]) -> int:
    # Silhouette expects Euclidean; OK in tangent space.
    best_k, best_s = None, -1.0
    for k in k_range:
        if k <= 1 or k >= T.shape[0]:
            continue
        labels_k = cluster_in_tangent_space(T, method="bisectkmeans", k=k)
        # If degenerate (1 cluster), skip
        if len(np.unique(labels_k)) < 2:
            continue
        s = silhouette_score(T, labels_k, metric="euclidean")
        if s > best_s:
            best_s, best_k = s, k
    return best_k if best_k is not None else N_CLUSTERS

# ------------------------------------------------------------
# 15) Run per bin
# ------------------------------------------------------------
for (a, b) in LAYER_BINS:
    bn = bin_name(a, b)
    Vz = BIN_CENTROIDS_Z[bn]

    # Spread in D-dim Poincaré ball
    alpha_bin = choose_alpha_for_target_radius(Vz, c=C, r_target=R_TARGET_BALL)
    X_ball = hyperbolic_points_from_vectors(Vz, alpha_bin, C)

    radii = np.linalg.norm(X_ball, axis=1)
    print(f"\n=== Bin {bn} ===")
    print(f"alpha_bin={alpha_bin:.5g} | r_med={np.median(radii):.3f} | r_max={np.max(radii):.3f} | sat@0.95={sat_fraction(X_ball, C):.3f}")

    # Tangent representation for clustering
    T = logmap0(X_ball, c=C, eps=EPS)  # (N,D)

    # Optional PCA for clustering stability/speed (does not affect hyperbolic mapping)
    d = min(TANGENT_PCA_DIM_FOR_CLUSTERING, T.shape[1])
    T_red = PCA(n_components=d, random_state=0).fit_transform(T)

    # Choose k (optional)
    if AUTO_K:
        k_bin = choose_k_by_silhouette(T_red, K_RANGE)
    else:
        k_bin = N_CLUSTERS

    # Cluster ALL languages (balanced)
    clusters = cluster_in_tangent_space(T_red, method=CLUSTER_METHOD, k=k_bin)

    uniq, cnt = np.unique(clusters, return_counts=True)
    print(f"Clustering: method={CLUSTER_METHOD}, k={k_bin}")
    print("Cluster sizes:", {int(u): int(c_) for u, c_ in zip(uniq, cnt)})

    if PRINT_CLUSTER_LISTS:
        print_cluster_lists(bn, clusters, META, max_per_cluster=None)

    # Distances only for edge induction (visualization)
    D = pairwise_poincare_distance_matrix(X_ball, c=C)
    if EDGE_MODE == "mst":
        edges = edges_from_mst(D)
    elif EDGE_MODE == "knn":
        edges = edges_from_knn(D, k=KNN_K)
    else:
        edges = None

    # 2D disk coords for visualization
    X2 = poincare_disk_coords_scaled(X_ball, c=C, r_target=R_TARGET_VIS)
    X2_centered = mobius_translate_to_origin(X2, X2[fo_idx], c=C)

    # Normalize to unit disk for plotting
    R = 1.0 / math.sqrt(C)
    X2_unit = X2_centered / max(R, 1e-12)

    # Label indices for plot (names only)
    show_labels = (LABEL_MODE != "none")
    label_indices = None
    if LABEL_MODE == "subset":
        label_indices = select_label_indices(clusters, fo_idx, top_k_per_cluster=LABEL_TOP_K_PER_CLUSTER)
    elif LABEL_MODE == "all":
        label_indices = list(range(len(labels_name_only)))

    plot_poincare_disk_graph(
        coords2d_unit=X2_unit,
        clusters=clusters,
        fo_idx=fo_idx,
        title=f"Poincaré disk graph (bin {bn}) | C={C}, alpha={alpha_bin:.3g} | {CLUSTER_METHOD}, k={k_bin}",
        edges=edges,
        labels=labels_name_only,          # NAME ONLY on plot
        label_indices=label_indices,
        show_labels=show_labels,
        epoch_text=f"Bin: {bn}",
    )
```

```python
# ============================================================
# Colab: Hyperbolic clustering + Poincaré disk visualization
# (WordNet-style disk with induced edges + cluster listings)
#
# FIXED: Correct layer handling for (L, D) arrays in .npz
# - Prevents accidental averaging across layers
# - Ensures bins differ when layer representations differ
#
# Features:
# - Loads per-language embeddings from .npz on Google Drive
# - Builds per-layer language centroids (sentence-mean if needed)
# - Bins layers: [0-4], [5-9], [10-14], [15-19], [20-24]
# - Projects to Poincaré ball via expmap0 with auto alpha (target median radius)
# - Clusters ALL languages in tangent space using BisectingKMeans or Ward
# - Visualizes Poincaré disk (2D via PCA in tangent space, mapped back)
# - Centers Faroese for visualization only (Möbius translation)
# - Draws MST (or kNN) edges using hyperbolic distances
# - Labels on plot are NAME ONLY; printed cluster lists include codes
# ============================================================

!pip -q install scipy umap-learn adjustText

import os, re, glob, math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection

from google.colab import drive
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering, BisectingKMeans
from sklearn.metrics import silhouette_score
from scipy.sparse.csgraph import minimum_spanning_tree

# -----------------------
# 1) User config
# -----------------------
BASE_DIR = "/content/drive/MyDrive/Phd/datasets/wav2vec2_embeddings/paper1"
NPZ_GLOB = os.path.join(BASE_DIR, "*.npz")

LAYER_BINS = [(0,4), (5,9), (10,14), (15,19), (20,24)]
ALL_LAYERS = list(range(0, 25))

# Hyperbolic parameters
C = 1.0
R_TARGET_BALL = 0.60
R_TARGET_VIS = 0.60

# --- Clustering config ---
CLUSTER_METHOD = "bisectkmeans"   # "bisectkmeans" | "ward"
N_CLUSTERS = 10
AUTO_K = False
K_RANGE = list(range(4, 21))
TANGENT_PCA_DIM_FOR_CLUSTERING = 32  # tangent PCA dim used only for clustering stability/speed

# Graph edges
EDGE_MODE = "mst"  # "mst" | "knn" | "none"
KNN_K = 3

# Plotting
FIGSIZE = (10, 10)
POINT_SIZE = 18
EDGE_ALPHA = 0.18
EDGE_LW = 0.20

# Label options (names only)
LABEL_MODE = "subset"     # "none" | "subset" | "all"
LABEL_TOP_K_PER_CLUSTER = 6
LABEL_FONT_SIZE = 8

PRINT_CLUSTER_LISTS = True

EPS = 1e-9
SAFE_EPS_BALL = 1e-5

# -----------------------
# 2) Mount Drive
# -----------------------
drive.mount("/content/drive")

# ------------------------------------------------------------
# 3) Filename parsing
# ------------------------------------------------------------
LANG_RE = re.compile(r"embeddings_language-(.+?)_languagecode-\[([^\]]+)\]")

def parse_lang_from_filename(path: str):
    fn = os.path.basename(path)
    m = LANG_RE.search(fn)
    if m:
        return m.group(1), m.group(2)
    return os.path.splitext(fn)[0], "unknown"

# ------------------------------------------------------------
# 4) Robust layer extractor (FIXED for (L,D))
# ------------------------------------------------------------
def get_layer_matrix(npz: np.lib.npyio.NpzFile, layer: int) -> np.ndarray:
    """
    Returns a (N, D) matrix for the requested layer.
    If the file stores per-layer centroids as (L, D), returns (1, D).
    """
    keys = list(npz.keys())

    # Case A: single key holding embeddings
    for k in ["embeddings", "X", "hidden_states", "layer_embeddings"]:
        if k in keys:
            arr = np.asarray(npz[k])

            # (L, N, D) -> (N, D)
            if arr.ndim == 3 and arr.shape[0] > layer:
                return arr[layer]

            # (N, L, D) -> (N, D)
            if arr.ndim == 3 and arr.shape[1] > layer:
                return arr[:, layer, :]

            # CRITICAL FIX:
            # (L, D) already-averaged per layer -> (1, D) for layer
            if arr.ndim == 2 and arr.shape[0] >= 25 and arr.shape[1] >= 64:
                if arr.shape[0] > layer:
                    return arr[layer][None, :]

            # Transposed (D, L) -> (1, D)
            if arr.ndim == 2 and arr.shape[1] >= 25 and arr.shape[0] >= 64:
                if arr.shape[1] > layer:
                    return arr[:, layer][None, :]

            # Sentence matrix (N, D) -> return as-is
            if arr.ndim == 2:
                return arr

            raise ValueError(f"Unrecognized shape for key '{k}': {arr.shape}")

    # Case B: per-layer keys
    candidates = [
        f"embeddings_layer_{layer}",
        f"layer_{layer}",
        f"hidden_state_{layer}",
        f"h_{layer}",
        f"layer{layer}",
    ]
    for ck in candidates:
        if ck in keys:
            arr = np.asarray(npz[ck])
            if arr.ndim == 1:
                return arr[None, :]
            if arr.ndim == 2:
                return arr
            raise ValueError(f"Key '{ck}' exists but has shape={arr.shape}")

    # Case C: last resort key match
    layer_pat = re.compile(rf"(^|[^0-9]){layer}([^0-9]|$)")
    for k in keys:
        if layer_pat.search(k):
            arr = np.asarray(npz[k])
            if arr.ndim == 1:
                return arr[None, :]
            if arr.ndim == 2:
                return arr

    raise KeyError(f"Could not find embeddings for layer={layer}. keys={keys[:40]}")

# ------------------------------------------------------------
# 5) Load language: centroid per layer (FIXED)
# ------------------------------------------------------------
def load_language(npz_path: str, layers=ALL_LAYERS):
    lang_name, lang_code = parse_lang_from_filename(npz_path)
    z = np.load(npz_path, allow_pickle=True)

    centroids = {}
    for l in layers:
        X = get_layer_matrix(z, l)
        X = np.asarray(X, dtype=np.float32)

        # If already a per-layer centroid -> (1, D)
        if X.ndim == 2 and X.shape[0] == 1:
            centroids[l] = X[0]
        # If sentence matrix -> (N, D)
        elif X.ndim == 2:
            centroids[l] = X.mean(axis=0)
        else:
            raise ValueError(f"Unexpected shape for layer {l}: {X.shape}")

    z.close()
    return {"path": npz_path, "lang_name": lang_name, "lang_code": lang_code, "centroids": centroids}

paths = sorted(glob.glob(NPZ_GLOB))
if not paths:
    raise FileNotFoundError(f"No .npz files found under: {BASE_DIR}")

languages = []
for p in paths:
    try:
        languages.append(load_language(p))
    except Exception as e:
        print(f"[WARN] Skipping {os.path.basename(p)} due to: {e}")

print(f"Loaded {len(languages)} languages.")

# ------------------------------------------------------------
# 6) Identify Faroese (visualization centering only)
# ------------------------------------------------------------
def is_faroese(lang_name, lang_code):
    ln = (lang_name or "").lower()
    lc = (lang_code or "").lower().strip()
    return ("faro" in ln) or (lc in ["fo", "fo_fo", "fo-fo"]) or ("fao" in ln)

fo_candidates = [x for x in languages if is_faroese(x["lang_name"], x["lang_code"])]
if not fo_candidates:
    raise ValueError("Could not identify Faroese file. Adjust is_faroese().")
FO = fo_candidates[0]
print("Using Faroese:", FO["lang_name"], FO["lang_code"], os.path.basename(FO["path"]))

# Sanity check: Faroese should vary across layers
tmp = np.stack([FO["centroids"][l] for l in range(25)], axis=0)
print("SANITY: FO layer std (mean over dims):", float(tmp.std(axis=0).mean()))

# ------------------------------------------------------------
# 7) Hyperbolic primitives
# ------------------------------------------------------------
def expmap0(v: np.ndarray, c: float, eps: float = 1e-9) -> np.ndarray:
    v = np.asarray(v, dtype=np.float64)
    norm = np.linalg.norm(v, axis=-1, keepdims=True)
    norm = np.clip(norm, eps, None)
    sqrt_c = math.sqrt(c)
    z = np.clip(sqrt_c * norm, 0.0, 20.0)
    factor = np.tanh(z) / np.clip(z, eps, None)
    return factor * v

def logmap0(x: np.ndarray, c: float, eps: float = 1e-9) -> np.ndarray:
    x = np.asarray(x, dtype=np.float64)
    norm = np.linalg.norm(x, axis=-1, keepdims=True)
    norm = np.clip(norm, eps, None)
    sqrt_c = math.sqrt(c)
    z = np.clip(sqrt_c * norm, 0.0, 1.0 - 1e-6)
    factor = np.arctanh(z) / (sqrt_c * norm)
    return factor * x

def poincare_dist(x: np.ndarray, y: np.ndarray, c: float, eps: float = 1e-9) -> float:
    x = np.asarray(x, dtype=np.float64)
    y = np.asarray(y, dtype=np.float64)
    x2 = float(np.dot(x, x))
    y2 = float(np.dot(y, y))
    max_r2 = (1.0 / c) - eps
    x2 = min(x2, max_r2)
    y2 = min(y2, max_r2)

    diff2 = float(np.dot(x - y, x - y))
    denom = (1.0 - c * x2) * (1.0 - c * y2)
    denom = max(denom, eps)

    z = 1.0 + (2.0 * c * diff2) / denom
    z = max(z, 1.0 + eps)
    return (1.0 / math.sqrt(c)) * math.acosh(z)

def mobius_add(x: np.ndarray, y: np.ndarray, c: float, eps: float = 1e-9) -> np.ndarray:
    x = np.asarray(x, dtype=np.float64)
    y = np.asarray(y, dtype=np.float64)
    x2 = np.sum(x*x, axis=-1, keepdims=True)
    y2 = np.sum(y*y, axis=-1, keepdims=True)
    xy = np.sum(x*y, axis=-1, keepdims=True)
    num = (1 + 2*c*xy + c*y2) * x + (1 - c*x2) * y
    den = 1 + 2*c*xy + (c**2) * x2 * y2
    den = np.clip(den, eps, None)
    return num / den

def mobius_translate_to_origin(X: np.ndarray, x0: np.ndarray, c: float) -> np.ndarray:
    return mobius_add(-x0, X, c=c, eps=EPS)

# ------------------------------------------------------------
# 8) Auto-scaling
# ------------------------------------------------------------
def choose_alpha_for_target_radius(Vz: np.ndarray, c: float, r_target: float = 0.6, eps: float = 1e-9) -> float:
    norms = np.linalg.norm(Vz.astype(np.float64), axis=1)
    med = float(np.median(norms))
    med = max(med, eps)
    sqrt_c = math.sqrt(c)
    r_cap = min(r_target, (1.0 - 1e-5) / sqrt_c)
    t_target = (1.0 / sqrt_c) * math.atanh(r_cap * sqrt_c)
    return t_target / med

def sat_fraction(X_ball: np.ndarray, c: float, thresh: float = 0.95) -> float:
    R = 1.0 / math.sqrt(c)
    r = np.linalg.norm(X_ball, axis=1)
    return float(np.mean(r >= thresh * R))

def hyperbolic_points_from_vectors(Vz: np.ndarray, alpha: float, c: float) -> np.ndarray:
    X = expmap0(alpha * Vz, c=c, eps=EPS)
    r2 = np.sum(X*X, axis=1)
    rmax = (1.0 - SAFE_EPS_BALL)**2 / c
    too_big = r2 >= rmax
    if np.any(too_big):
        scale = np.sqrt(rmax / np.clip(r2[too_big], 1e-12, None))
        X[too_big] *= scale[:, None]
    return X

def poincare_disk_coords_scaled(X_ball: np.ndarray, c: float, r_target: float = 0.6) -> np.ndarray:
    T = logmap0(X_ball, c=c, eps=EPS)
    T2 = PCA(n_components=2, random_state=0).fit_transform(T)
    norms2 = np.linalg.norm(T2, axis=1)
    med2 = float(np.median(norms2))
    med2 = max(med2, 1e-9)
    sqrt_c = math.sqrt(c)
    r_cap = min(r_target, (1.0 - 1e-5) / sqrt_c)
    t_target = (1.0 / sqrt_c) * math.atanh(r_cap * sqrt_c)
    vis_scale = t_target / med2
    return expmap0(T2 * vis_scale, c=c, eps=EPS)

# ------------------------------------------------------------
# 9) Poincaré distance matrix (for edges)
# ------------------------------------------------------------
def pairwise_poincare_distance_matrix(X_ball: np.ndarray, c: float) -> np.ndarray:
    n = X_ball.shape[0]
    D = np.zeros((n, n), dtype=np.float64)
    for i in range(n):
        for j in range(i+1, n):
            d = poincare_dist(X_ball[i], X_ball[j], c=c, eps=EPS)
            D[i, j] = d
            D[j, i] = d
    return D

# ------------------------------------------------------------
# 10) Build bin centroids and standardize per bin across languages
# ------------------------------------------------------------
def bin_name(a, b): return f"L{a:02d}-{b:02d}"

BIN_CENTROIDS = {bin_name(a,b): [] for (a,b) in LAYER_BINS}
META = []  # (name, code, path)

for entry in languages:
    META.append((entry["lang_name"], entry["lang_code"], entry["path"]))
    for (a,b) in LAYER_BINS:
        mus = np.stack([entry["centroids"][l] for l in range(a, b+1)], axis=0)
        BIN_CENTROIDS[bin_name(a,b)].append(mus.mean(axis=0))

for k in BIN_CENTROIDS:
    BIN_CENTROIDS[k] = np.stack(BIN_CENTROIDS[k], axis=0).astype(np.float32)

BIN_CENTROIDS_Z = {}
for k, M in BIN_CENTROIDS.items():
    BIN_CENTROIDS_Z[k] = StandardScaler().fit_transform(M).astype(np.float32)

# Faroese index (for visualization centering only)
fo_idx_list = [i for i,(n,code,_) in enumerate(META) if code == FO["lang_code"]]
if not fo_idx_list:
    raise ValueError("Faroese not found after loading; check parsing.")
fo_idx = fo_idx_list[0]

labels_name_only = [n for (n,_,_) in META]

# ------------------------------------------------------------
# 11) Induced edges
# ------------------------------------------------------------
def edges_from_knn(D: np.ndarray, k: int = 3):
    N = D.shape[0]
    edges = set()
    for i in range(N):
        nn = np.argpartition(D[i], kth=min(k+1, N-1))[:k+1]
        nn = [j for j in nn if j != i][:k]
        for j in nn:
            a, b = (i, j) if i < j else (j, i)
            edges.add((a, b))
    return sorted(edges)

def edges_from_mst(D: np.ndarray):
    T = minimum_spanning_tree(D).tocoo()
    edges = set()
    for i, j in zip(T.row, T.col):
        a, b = (int(i), int(j)) if i < j else (int(j), int(i))
        edges.add((a, b))
    return sorted(edges)

# ------------------------------------------------------------
# 12) Cluster printing + label selection
# ------------------------------------------------------------
def print_cluster_lists(bin_str: str, clusters: np.ndarray, meta: list, max_per_cluster=None):
    print(f"\n--- Cluster membership (bin {bin_str}) ---")
    for cid in sorted(np.unique(clusters).tolist()):
        idx = np.where(clusters == cid)[0].tolist()
        items = [f"{meta[i][0]} [{meta[i][1]}]" for i in idx]
        if max_per_cluster is not None and len(items) > max_per_cluster:
            head = items[:max_per_cluster]
            print(f"Cluster {cid:>2}: n={len(items):>3} | " + ", ".join(head) + f", ... (+{len(items)-max_per_cluster})")
        else:
            print(f"Cluster {cid:>2}: n={len(items):>3} | " + ", ".join(items))

def select_label_indices(clusters: np.ndarray, fo_idx: int, top_k_per_cluster: int = 6):
    keep = {fo_idx}
    for cid in sorted(np.unique(clusters).tolist()):
        idx = np.where(clusters == cid)[0]
        for i in idx[:top_k_per_cluster]:
            keep.add(int(i))
    return sorted(keep)

# ------------------------------------------------------------
# 13) Plot
# ------------------------------------------------------------
def plot_poincare_disk_graph(coords2d_unit, clusters, fo_idx, title, edges=None,
                             labels=None, label_indices=None, show_labels=False,
                             epoch_text=None):
    coords2d_unit = np.asarray(coords2d_unit)
    fig, ax = plt.subplots(figsize=FIGSIZE)
    ax.set_aspect("equal")

    ax.add_patch(
        plt.Circle((0,0), 1.0, fill=True, ec="black", fc="#f8f9f9",
                   linestyle="-.", linewidth=1.0, zorder=0)
    )

    if edges is not None and len(edges) > 0:
        segs = [(coords2d_unit[i], coords2d_unit[j]) for (i, j) in edges]
        lc = LineCollection(segs, colors="blue", linewidths=EDGE_LW, alpha=EDGE_ALPHA, zorder=1)
        ax.add_collection(lc)

    ax.scatter(coords2d_unit[:,0], coords2d_unit[:,1],
               s=POINT_SIZE, c=clusters, alpha=0.90, zorder=2)

    ax.scatter([coords2d_unit[fo_idx,0]], [coords2d_unit[fo_idx,1]],
               marker="*", s=320, c="red", zorder=3, label="Faroese (centered)")

    if show_labels and labels is not None:
        if label_indices is None:
            label_indices = list(range(len(labels)))
        for i in label_indices:
            ax.text(coords2d_unit[i,0], coords2d_unit[i,1], labels[i],
                    fontsize=LABEL_FONT_SIZE, zorder=4)

    if epoch_text is not None:
        ax.text(-0.98, 0.90, epoch_text, fontsize=18, weight="bold", alpha=0.6)

    ax.set_title(title)
    ax.set_xlim([-1.05, 1.05])
    ax.set_ylim([-1.05, 1.05])
    ax.axis("off")
    ax.legend(loc="upper right")
    plt.show()

# ------------------------------------------------------------
# 14) Clustering in tangent space
# ------------------------------------------------------------
def cluster_in_tangent_space(T: np.ndarray, method: str, k: int) -> np.ndarray:
    if method == "bisectkmeans":
        model = BisectingKMeans(n_clusters=k, random_state=0)
        return model.fit_predict(T)
    if method == "ward":
        model = AgglomerativeClustering(n_clusters=k, linkage="ward")
        return model.fit_predict(T)
    raise ValueError(f"Unknown method: {method}")

def choose_k_by_silhouette(T: np.ndarray, k_range):
    best_k, best_s = None, -1.0
    for k in k_range:
        if k <= 1 or k >= T.shape[0]:
            continue
        labels_k = cluster_in_tangent_space(T, method="bisectkmeans", k=k)
        if len(np.unique(labels_k)) < 2:
            continue
        s = silhouette_score(T, labels_k, metric="euclidean")
        if s > best_s:
            best_s, best_k = s, k
    return best_k if best_k is not None else N_CLUSTERS

# ------------------------------------------------------------
# 15) Run per bin
# ------------------------------------------------------------
for (a, b) in LAYER_BINS:
    bn = bin_name(a, b)
    Vz = BIN_CENTROIDS_Z[bn]

    alpha_bin = choose_alpha_for_target_radius(Vz, c=C, r_target=R_TARGET_BALL)
    X_ball = hyperbolic_points_from_vectors(Vz, alpha_bin, C)

    radii = np.linalg.norm(X_ball, axis=1)
    print(f"\n=== Bin {bn} ===")
    print(f"alpha_bin={alpha_bin:.5g} | r_med={np.median(radii):.3f} | r_max={np.max(radii):.3f} | sat@0.95={sat_fraction(X_ball, C):.3f}")

    # Tangent representation for clustering
    T = logmap0(X_ball, c=C, eps=EPS)

    # PCA for clustering stability/speed
    d = min(TANGENT_PCA_DIM_FOR_CLUSTERING, T.shape[1])
    T_red = PCA(n_components=d, random_state=0).fit_transform(T)

    if AUTO_K:
        k_bin = choose_k_by_silhouette(T_red, K_RANGE)
    else:
        k_bin = N_CLUSTERS

    clusters = cluster_in_tangent_space(T_red, method=CLUSTER_METHOD, k=k_bin)

    uniq, cnt = np.unique(clusters, return_counts=True)
    print(f"Clustering: method={CLUSTER_METHOD}, k={k_bin}")
    print("Cluster sizes:", {int(u): int(c_) for u, c_ in zip(uniq, cnt)})

    if PRINT_CLUSTER_LISTS:
        print_cluster_lists(bn, clusters, META, max_per_cluster=None)

    # Distances for edge induction
    D = pairwise_poincare_distance_matrix(X_ball, c=C)

    if EDGE_MODE == "mst":
        edges = edges_from_mst(D)
    elif EDGE_MODE == "knn":
        edges = edges_from_knn(D, k=KNN_K)
    elif EDGE_MODE == "none":
        edges = None
    else:
        raise ValueError("EDGE_MODE must be 'mst', 'knn', or 'none'.")

    # 2D disk coords for visualization
    X2 = poincare_disk_coords_scaled(X_ball, c=C, r_target=R_TARGET_VIS)

    # Center Faroese for visualization
    X2_centered = mobius_translate_to_origin(X2, X2[fo_idx], c=C)

    # Normalize to unit disk for plotting
    R = 1.0 / math.sqrt(C)
    X2_unit = X2_centered / max(R, 1e-12)

    show_labels = (LABEL_MODE != "none")
    if LABEL_MODE == "subset":
        label_indices = select_label_indices(clusters, fo_idx, top_k_per_cluster=LABEL_TOP_K_PER_CLUSTER)
    elif LABEL_MODE == "all":
        label_indices = list(range(len(labels_name_only)))
    else:
        label_indices = None

    plot_poincare_disk_graph(
        coords2d_unit=X2_unit,
        clusters=clusters,
        fo_idx=fo_idx,
        title=f"Poincaré disk graph (bin {bn}) | C={C}, alpha={alpha_bin:.3g} | {CLUSTER_METHOD}, k={k_bin}",
        edges=edges,
        #labels=labels_name_only,
        label_indices=label_indices,
        show_labels=show_labels,
        epoch_text=f"Bin: {bn}",
    )
```

```python
# ============================================================
# Recompute + save ONE composite figure (5 Poincaré disks)
# - One subplot per layer bin (0-4, 5-9, 10-14, 15-19, 20-24)
# - NO per-plot titles (paper-ready)
# - Faroese centered (red star), MST edges, cluster colors
# - Saves: poincare_disks_5bins.pdf and .png
# ============================================================

!pip -q install scipy

import os, re, glob, math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection

from google.colab import drive
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import BisectingKMeans
from scipy.sparse.csgraph import minimum_spanning_tree

# -----------------------
# Config
# -----------------------
BASE_DIR = "/content/drive/MyDrive/Phd/datasets/wav2vec2_embeddings/paper1"
NPZ_GLOB = os.path.join(BASE_DIR, "*.npz")

LAYER_BINS = [(0,4), (5,9), (10,14), (15,19), (20,24)]
ALL_LAYERS = list(range(0, 25))

# Hyperbolic
C = 1.0
R_TARGET_BALL = 0.60
R_TARGET_VIS = 0.60
EPS = 1e-9
SAFE_EPS_BALL = 1e-5

# Clustering
N_CLUSTERS = 10
TANGENT_PCA_DIM_FOR_CLUSTERING = 32

# Edges
EDGE_MODE = "mst"  # "mst" only for paper composite
EDGE_ALPHA = 0.18
EDGE_LW = 0.25

# Plot style
FIGSIZE = (14, 8)         # adjust for paper layout
POINT_SIZE = 18
STAR_SIZE = 260
SHOW_LEGEND = False       # paper: usually off
BIN_LABELS = True         # show small bin labels in corners

# Output
OUT_PDF = "poincare_disks_5bins.pdf"
OUT_PNG = "poincare_disks_5bins.png"
DPI = 300

# -----------------------
# Mount Drive
# -----------------------
drive.mount("/content/drive")

# -----------------------
# Filename parsing
# -----------------------
LANG_RE = re.compile(r"embeddings_language-(.+?)_languagecode-\[([^\]]+)\]")

def parse_lang_from_filename(path: str):
    fn = os.path.basename(path)
    m = LANG_RE.search(fn)
    if m:
        return m.group(1), m.group(2)
    return os.path.splitext(fn)[0], "unknown"

# -----------------------
# Robust layer extractor (FIXED for (L,D))
# -----------------------
def get_layer_matrix(npz: np.lib.npyio.NpzFile, layer: int) -> np.ndarray:
    keys = list(npz.keys())

    for k in ["embeddings", "X", "hidden_states", "layer_embeddings"]:
        if k in keys:
            arr = np.asarray(npz[k])

            # (L, N, D)
            if arr.ndim == 3 and arr.shape[0] > layer:
                return arr[layer]

            # (N, L, D)
            if arr.ndim == 3 and arr.shape[1] > layer:
                return arr[:, layer, :]

            # (L, D) layer centroids
            if arr.ndim == 2 and arr.shape[0] >= 25 and arr.shape[1] >= 64:
                if arr.shape[0] > layer:
                    return arr[layer][None, :]

            # (D, L) transposed layer centroids
            if arr.ndim == 2 and arr.shape[1] >= 25 and arr.shape[0] >= 64:
                if arr.shape[1] > layer:
                    return arr[:, layer][None, :]

            # (N, D) sentence matrix (single layer)
            if arr.ndim == 2:
                return arr

            raise ValueError(f"Unrecognized shape for key '{k}': {arr.shape}")

    # per-layer keys
    candidates = [
        f"embeddings_layer_{layer}",
        f"layer_{layer}",
        f"hidden_state_{layer}",
        f"h_{layer}",
        f"layer{layer}",
    ]
    for ck in candidates:
        if ck in keys:
            arr = np.asarray(npz[ck])
            if arr.ndim == 1:
                return arr[None, :]
            if arr.ndim == 2:
                return arr
            raise ValueError(f"Key '{ck}' exists but has shape={arr.shape}")

    # last resort key match
    layer_pat = re.compile(rf"(^|[^0-9]){layer}([^0-9]|$)")
    for k in keys:
        if layer_pat.search(k):
            arr = np.asarray(npz[k])
            if arr.ndim == 1:
                return arr[None, :]
            if arr.ndim == 2:
                return arr

    raise KeyError(f"Could not find embeddings for layer={layer}. keys={keys[:40]}")

# -----------------------
# Load language centroids per layer
# -----------------------
def load_language(npz_path: str, layers=ALL_LAYERS):
    lang_name, lang_code = parse_lang_from_filename(npz_path)
    z = np.load(npz_path, allow_pickle=True)

    centroids = {}
    for l in layers:
        X = get_layer_matrix(z, l)
        X = np.asarray(X, dtype=np.float32)

        if X.ndim == 2 and X.shape[0] == 1:
            centroids[l] = X[0]
        elif X.ndim == 2:
            centroids[l] = X.mean(axis=0)
        else:
            raise ValueError(f"Unexpected shape for layer {l}: {X.shape}")

    z.close()
    return {"path": npz_path, "lang_name": lang_name, "lang_code": lang_code, "centroids": centroids}

paths = sorted(glob.glob(NPZ_GLOB))
if not paths:
    raise FileNotFoundError(f"No .npz files found under: {BASE_DIR}")

languages = []
for p in paths:
    try:
        languages.append(load_language(p))
    except Exception as e:
        print(f"[WARN] Skipping {os.path.basename(p)} due to: {e}")

print(f"Loaded {len(languages)} languages.")

# -----------------------
# Identify Faroese index
# -----------------------
def is_faroese(lang_name, lang_code):
    ln = (lang_name or "").lower()
    lc = (lang_code or "").lower().strip()
    return ("faro" in ln) or (lc in ["fo", "fo_fo", "fo-fo"]) or ("fao" in ln)

fo_candidates = [x for x in languages if is_faroese(x["lang_name"], x["lang_code"])]
if not fo_candidates:
    raise ValueError("Could not identify Faroese file. Adjust is_faroese().")
FO = fo_candidates[0]

META = [(x["lang_name"], x["lang_code"], x["path"]) for x in languages]
fo_idx_list = [i for i,(n,code,_) in enumerate(META) if code == FO["lang_code"]]
if not fo_idx_list:
    raise ValueError("Faroese not found after loading; check parsing.")
fo_idx = fo_idx_list[0]

# -----------------------
# Hyperbolic primitives
# -----------------------
def expmap0(v: np.ndarray, c: float, eps: float = 1e-9) -> np.ndarray:
    v = np.asarray(v, dtype=np.float64)
    norm = np.linalg.norm(v, axis=-1, keepdims=True)
    norm = np.clip(norm, eps, None)
    sqrt_c = math.sqrt(c)
    z = np.clip(sqrt_c * norm, 0.0, 20.0)
    factor = np.tanh(z) / np.clip(z, eps, None)
    return factor * v

def logmap0(x: np.ndarray, c: float, eps: float = 1e-9) -> np.ndarray:
    x = np.asarray(x, dtype=np.float64)
    norm = np.linalg.norm(x, axis=-1, keepdims=True)
    norm = np.clip(norm, eps, None)
    sqrt_c = math.sqrt(c)
    z = np.clip(sqrt_c * norm, 0.0, 1.0 - 1e-6)
    factor = np.arctanh(z) / (sqrt_c * norm)
    return factor * x

def poincare_dist(x: np.ndarray, y: np.ndarray, c: float, eps: float = 1e-9) -> float:
    x = np.asarray(x, dtype=np.float64)
    y = np.asarray(y, dtype=np.float64)
    x2 = float(np.dot(x, x))
    y2 = float(np.dot(y, y))
    max_r2 = (1.0 / c) - eps
    x2 = min(x2, max_r2)
    y2 = min(y2, max_r2)
    diff2 = float(np.dot(x - y, x - y))
    denom = (1.0 - c * x2) * (1.0 - c * y2)
    denom = max(denom, eps)
    z = 1.0 + (2.0 * c * diff2) / denom
    z = max(z, 1.0 + eps)
    return (1.0 / math.sqrt(c)) * math.acosh(z)

def mobius_add(x: np.ndarray, y: np.ndarray, c: float, eps: float = 1e-9) -> np.ndarray:
    x = np.asarray(x, dtype=np.float64)
    y = np.asarray(y, dtype=np.float64)
    x2 = np.sum(x*x, axis=-1, keepdims=True)
    y2 = np.sum(y*y, axis=-1, keepdims=True)
    xy = np.sum(x*y, axis=-1, keepdims=True)
    num = (1 + 2*c*xy + c*y2) * x + (1 - c*x2) * y
    den = 1 + 2*c*xy + (c**2) * x2 * y2
    den = np.clip(den, eps, None)
    return num / den

def mobius_translate_to_origin(X: np.ndarray, x0: np.ndarray, c: float) -> np.ndarray:
    return mobius_add(-x0, X, c=c, eps=EPS)

# -----------------------
# Scaling + helpers
# -----------------------
def choose_alpha_for_target_radius(Vz: np.ndarray, c: float, r_target: float = 0.6, eps: float = 1e-9) -> float:
    norms = np.linalg.norm(Vz.astype(np.float64), axis=1)
    med = float(np.median(norms))
    med = max(med, eps)
    sqrt_c = math.sqrt(c)
    r_cap = min(r_target, (1.0 - 1e-5) / sqrt_c)
    t_target = (1.0 / sqrt_c) * math.atanh(r_cap * sqrt_c)
    return t_target / med

def hyperbolic_points_from_vectors(Vz: np.ndarray, alpha: float, c: float) -> np.ndarray:
    X = expmap0(alpha * Vz, c=c, eps=EPS)
    r2 = np.sum(X*X, axis=1)
    rmax = (1.0 - SAFE_EPS_BALL)**2 / c
    too_big = r2 >= rmax
    if np.any(too_big):
        scale = np.sqrt(rmax / np.clip(r2[too_big], 1e-12, None))
        X[too_big] *= scale[:, None]
    return X

def poincare_disk_coords_scaled(X_ball: np.ndarray, c: float, r_target: float = 0.6) -> np.ndarray:
    # PCA in tangent space -> expmap back for 2D Poincaré disk
    T = logmap0(X_ball, c=c, eps=EPS)
    T2 = PCA(n_components=2, random_state=0).fit_transform(T)
    norms2 = np.linalg.norm(T2, axis=1)
    med2 = float(np.median(norms2))
    med2 = max(med2, 1e-9)
    sqrt_c = math.sqrt(c)
    r_cap = min(r_target, (1.0 - 1e-5) / sqrt_c)
    t_target = (1.0 / sqrt_c) * math.atanh(r_cap * sqrt_c)
    vis_scale = t_target / med2
    return expmap0(T2 * vis_scale, c=c, eps=EPS)

def pairwise_poincare_distance_matrix(X_ball: np.ndarray, c: float) -> np.ndarray:
    n = X_ball.shape[0]
    D = np.zeros((n, n), dtype=np.float64)
    for i in range(n):
        for j in range(i+1, n):
            d = poincare_dist(X_ball[i], X_ball[j], c=c, eps=EPS)
            D[i, j] = d
            D[j, i] = d
    return D

def edges_from_mst(D: np.ndarray):
    T = minimum_spanning_tree(D).tocoo()
    edges = []
    for i, j in zip(T.row, T.col):
        edges.append((int(i), int(j)))
    return edges

# -----------------------
# Build per-bin standardized language vectors
# -----------------------
def bin_name(a, b): return f"L{a:02d}-{b:02d}"

BIN_MATS = {}
for (a,b) in LAYER_BINS:
    M = []
    for entry in languages:
        mus = np.stack([entry["centroids"][l] for l in range(a, b+1)], axis=0)
        M.append(mus.mean(axis=0))
    M = np.stack(M, axis=0).astype(np.float32)
    Mz = StandardScaler().fit_transform(M).astype(np.float32)
    BIN_MATS[bin_name(a,b)] = Mz

# -----------------------
# One-bin compute: X_ball, clusters, edges, disk coords (centered)
# -----------------------
def compute_bin_artifacts(Vz_bin: np.ndarray):
    # hyperbolic in D dims
    alpha = choose_alpha_for_target_radius(Vz_bin, c=C, r_target=R_TARGET_BALL)
    X_ball = hyperbolic_points_from_vectors(Vz_bin, alpha=alpha, c=C)

    # tangent for clustering
    T = logmap0(X_ball, c=C, eps=EPS)
    d = min(TANGENT_PCA_DIM_FOR_CLUSTERING, T.shape[1])
    T_red = PCA(n_components=d, random_state=0).fit_transform(T)

    clusters = BisectingKMeans(n_clusters=N_CLUSTERS, random_state=0).fit_predict(T_red)

    # edges from MST using hyperbolic distances
    D = pairwise_poincare_distance_matrix(X_ball, c=C)
    edges = edges_from_mst(D)

    # 2D disk coords, centered on Faroese for visualization
    X2 = poincare_disk_coords_scaled(X_ball, c=C, r_target=R_TARGET_VIS)
    X2_centered = mobius_translate_to_origin(X2, X2[fo_idx], c=C)

    # normalize to unit disk for plotting
    R = 1.0 / math.sqrt(C)
    X2_unit = X2_centered / max(R, 1e-12)

    return alpha, clusters, edges, X2_unit

# -----------------------
# Plot helper (no titles)
# -----------------------
def plot_disk_on_ax(ax, X2_unit, clusters, edges, fo_idx, bin_label=None):
    ax.set_aspect("equal")

    # disk boundary
    ax.add_patch(plt.Circle((0,0), 1.0, fill=False, ec="black",
                            linestyle="-.", linewidth=1.0, zorder=0))

    # edges
    if edges:
        segs = [(X2_unit[i], X2_unit[j]) for (i, j) in edges]
        lc = LineCollection(segs, colors="blue", linewidths=EDGE_LW, alpha=EDGE_ALPHA, zorder=1)
        ax.add_collection(lc)

    # points
    ax.scatter(X2_unit[:,0], X2_unit[:,1], s=POINT_SIZE, c=clusters, alpha=0.95, zorder=2)

    # Faroese
    ax.scatter([X2_unit[fo_idx,0]], [X2_unit[fo_idx,1]],
               marker="*", s=STAR_SIZE, c="red", zorder=3)

    # small bin label
    if BIN_LABELS and bin_label is not None:
        ax.text(0.02, 0.98, bin_label, transform=ax.transAxes,
                ha="left", va="top", fontsize=12, fontweight="bold")

    ax.set_xlim([-1.05, 1.05])
    ax.set_ylim([-1.05, 1.05])
    ax.axis("off")

# -----------------------
# Compute all bins and plot as one composite
# -----------------------
bin_keys = [bin_name(a,b) for (a,b) in LAYER_BINS]

fig, axes = plt.subplots(2, 3, figsize=FIGSIZE)
axes = axes.flatten()

for idx, bk in enumerate(bin_keys):
    alpha, clusters, edges, X2_unit = compute_bin_artifacts(BIN_MATS[bk])
    plot_disk_on_ax(axes[idx], X2_unit, clusters, edges, fo_idx, bin_label=bk)

# last subplot unused (2x3 grid, 5 plots)
axes[-1].axis("off")

plt.tight_layout()
plt.savefig(OUT_PDF, bbox_inches="tight")
plt.savefig(OUT_PNG, dpi=DPI, bbox_inches="tight")
plt.show()

print("Saved:", OUT_PDF, "and", OUT_PNG)
```

```python
# ============================================================
# ADDITIONS: Euclidean NN tables + hyperbolic cluster printouts
# (does NOT change the existing plotting/saving behavior)
# ============================================================

TOPK_EUCLID = 15  # Faroese -> top-15 nearest (Euclidean) per bin

def lang_label(i: int) -> str:
    name, code, _ = META[i]
    name = (name or "").strip()
    code = (code or "").strip()
    if name and code:
        return f"{name} ({code})"
    return name or code or f"lang_{i}"

def print_euclidean_topk_for_bin(bk: str, Vz_bin: np.ndarray, top_k: int = 15):
    # Euclidean in the same standardized bin space used as input to hyperbolic mapping
    fo_vec = Vz_bin[fo_idx]
    dists = np.linalg.norm(Vz_bin - fo_vec, axis=1)

    order = np.argsort(dists)
    order = [i for i in order if i != fo_idx][:top_k]

    print(f"\n=== {bk} | Euclidean: Faroese top-{top_k} nearest (L2 in standardized bin space) ===")
    print(f"{'rank':>4}  {'language':<40}  {'dist':>10}")
    for r, i in enumerate(order, start=1):
        print(f"{r:>4}  {lang_label(i):<40}  {dists[i]:>10.4f}")

def print_hyperbolic_clusters_for_bin(bk: str, clusters: np.ndarray):
    # Uses the already-computed hyperbolic clustering labels (per bin)
    print(f"\n=== {bk} | Hyperbolic clusters (BisectingKMeans on tangent-space PCA) ===")
    for k in sorted(np.unique(clusters).tolist()):
        idxs = np.where(clusters == k)[0].tolist()
        # Put Faroese first if it is in this cluster (readability)
        if fo_idx in idxs:
            idxs.remove(fo_idx)
            idxs = [fo_idx] + idxs
        members = ", ".join(lang_label(i) for i in idxs)
        print(f"Cluster {k:02d} (n={len(idxs)}): {members}")

# -----------------------
# Compute all bins and plot as one composite (UNCHANGED)
# + collect artifacts for printing
# -----------------------
bin_keys = [bin_name(a,b) for (a,b) in LAYER_BINS]

BIN_CLUSTERS = {}   # bk -> clusters
BIN_ALPHA = {}      # bk -> alpha (optional, for debugging)

fig, axes = plt.subplots(2, 3, figsize=FIGSIZE)
axes = axes.flatten()

for idx, bk in enumerate(bin_keys):
    alpha, clusters, edges, X2_unit = compute_bin_artifacts(BIN_MATS[bk])
    BIN_CLUSTERS[bk] = clusters
    BIN_ALPHA[bk] = alpha
    plot_disk_on_ax(axes[idx], X2_unit, clusters, edges, fo_idx, bin_label=bk)

# last subplot unused (2x3 grid, 5 plots)
axes[-1].axis("off")

plt.tight_layout()
plt.savefig(OUT_PDF, bbox_inches="tight")
plt.savefig(OUT_PNG, dpi=DPI, bbox_inches="tight")
plt.show()

print("Saved:", OUT_PDF, "and", OUT_PNG)

# -----------------------
# NEW: Print per-bin tables (Euclidean top-15) + hyperbolic clusters
# -----------------------
for bk in bin_keys:
    print_euclidean_topk_for_bin(bk, BIN_MATS[bk], top_k=TOPK_EUCLID)
    print_hyperbolic_clusters_for_bin(bk, BIN_CLUSTERS[bk])
```

```python
# ============================================================
# Euclidean 2D PCA (FO + 15 NN) — uniform frames + improved label placement
# - All subplots have identical frame sizes (match L05-09 look)
# - No per-panel auto-zoom to text bboxes (prevents bad label layouts)
# - Stronger, more stable adjustText configuration + leader lines
# - No saving to disk
# ============================================================

!pip -q install adjustText

import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from adjustText import adjust_text

EUCLID15_TOPK = 15
POINT_SIZE = 28
STAR_SIZE = 280
LABEL_FS = 8

def lang_name_only(i: int) -> str:
    name = (META[i][0] or "").strip()
    if name:
        return name
    code = (META[i][1] or "").strip()
    return code or f"lang_{i}"

def topk_indices_euclid(Vz_bin: np.ndarray, fo_idx: int, top_k: int = 15):
    fo_vec = Vz_bin[fo_idx]
    dists = np.linalg.norm(Vz_bin - fo_vec, axis=1)
    order = np.argsort(dists)
    order = [i for i in order if i != fo_idx][:top_k]
    return [fo_idx] + order

def compute_pca2_centered_subset(Vz_bin: np.ndarray, subset_idx: list, fo_idx: int):
    Vsub = Vz_bin[subset_idx]
    X2 = PCA(n_components=2, random_state=0).fit_transform(Vsub)
    fo_pos = subset_idx.index(fo_idx)
    X2c = X2 - X2[fo_pos]
    return X2c, fo_pos

def frame_axes(ax, lw=0.9, alpha=0.9):
    for spine in ax.spines.values():
        spine.set_visible(True)
        spine.set_linewidth(lw)
        spine.set_alpha(alpha)

def limits_from_points(X2c, pad=0.28):
    # same policy everywhere -> consistent zoom behaviour
    x = X2c[:, 0]; y = X2c[:, 1]
    xmin, xmax = float(x.min()), float(x.max())
    ymin, ymax = float(y.min()), float(y.max())
    xr = max(xmax - xmin, 1e-6)
    yr = max(ymax - ymin, 1e-6)
    xmin -= pad * xr; xmax += pad * xr
    ymin -= pad * yr; ymax += pad * yr
    return (xmin, xmax), (ymin, ymax)

def plot_panel(ax, X2c, subset_idx, fo_pos, bin_label=None):
    # IMPORTANT: do NOT force equal aspect; it changes frame shapes.
    ax.set_aspect("auto")

    xs, ys = X2c[:, 0], X2c[:, 1]

    # scatter points
    ax.scatter(xs, ys, s=POINT_SIZE, alpha=0.95, zorder=2)

    # Faroese star at origin
    ax.scatter([0.0], [0.0], marker="*", s=STAR_SIZE, c="red", zorder=3)

    # fixed limits from points (stable layout)
    (xmin, xmax), (ymin, ymax) = limits_from_points(X2c, pad=0.30)
    ax.set_xlim((xmin, xmax))
    ax.set_ylim((ymin, ymax))

    # bin label (corner)
    if BIN_LABELS and bin_label is not None:
        ax.text(0.02, 0.98, bin_label, transform=ax.transAxes,
                ha="left", va="top", fontsize=12, fontweight="bold", zorder=10)

    # labels: start slightly offset to reduce point collisions
    texts = []
    for j, lang_i in enumerate(subset_idx):
        if j == fo_pos:
            t = ax.text(0.0, 0.0, "Faroese",
                        fontsize=LABEL_FS + 1,
                        fontweight="bold",
                        zorder=6,
                        bbox=dict(boxstyle="round,pad=0.15", fc="white", ec="none", alpha=0.75))
        else:
            t = ax.text(xs[j], ys[j], lang_name_only(lang_i),
                        fontsize=LABEL_FS,
                        alpha=0.95,
                        zorder=4)
        texts.append(t)

    # label adjustment with consistent leader lines
    # Key changes vs before:
    # - no auto-zoom after adjustments
    # - allow moving in xy, but penalize excessive movement (pulling force)
    # - stronger separation between texts and points
    adjust_text(
        texts,
        x=xs, y=ys,
        ax=ax,
        only_move={'points':'', 'text':'xy'},
        # encourage separation
        expand_points=(2.4, 2.8),
        expand_text=(1.8, 2.1),
        # forces: keep labels apart but not too far
        force_points=0.25,
        force_text=0.90,
        force_pull=0.06,      # pulls labels back toward their original position
        pull_threshold=8,
        lim=800,
        precision=0.001,
        arrowprops=dict(arrowstyle="-", lw=0.7, alpha=0.65)
    )

    # cosmetics
    ax.set_xticks([]); ax.set_yticks([])
    frame_axes(ax)

# -----------------------
# Composite layout: identical subplot rectangles + tight spacing
# -----------------------
bin_keys = [bin_name(a, b) for (a, b) in LAYER_BINS]

fig, axes = plt.subplots(2, 3, figsize=(14, 8))
axes = axes.flatten()

for i, bk in enumerate(bin_keys):
    subset_idx = topk_indices_euclid(BIN_MATS[bk], fo_idx=fo_idx, top_k=EUCLID15_TOPK)
    X2c, fo_pos = compute_pca2_centered_subset(BIN_MATS[bk], subset_idx, fo_idx=fo_idx)
    plot_panel(axes[i], X2c, subset_idx, fo_pos, bin_label=bk)

# last subplot unused
axes[-1].axis("off")

# Make subplot frames uniform and closer together (match the "wide" look of L05-09)
fig.subplots_adjust(left=0.03, right=0.995, bottom=0.04, top=0.98, wspace=0.06, hspace=0.10)

plt.show()
```
